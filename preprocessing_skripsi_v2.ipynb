```python
def preprocess_text(text):
    if pd.isna(text):
        return ""

    # lowercase
    text = text.lower()

    # hapus url
    text = re.sub(r"http\S+|www\S+", " ", text)

    # hapus mention & hashtag (isi katanya dibuang karena opini biasanya di teks utama)
    text = re.sub(r"@\w+|#\w+", " ", text)

    # hapus emoji & simbol non-huruf
    text = re.sub(r"[^a-zA-Z\s]", " ", text)

    # normalisasi slang
    tokens = text.split()
    tokens = [NORMALIZATION_DICT.get(t, t) for t in tokens]

    # hapus token satu huruf (noise)
    tokens = [t for t in tokens if len(t) > 1]

    # gabung lagi
    text = " ".join(tokens)

    # hapus spasi berlebih
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df["ulasan_clean"] = df["ulasan"].apply(preprocess_text)
df[["ulasan", "ulasan_clean"]].head()
```
