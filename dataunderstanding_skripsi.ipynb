### Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import string
```

### Data Understanding / EDA

```python
df = pd.read_excel('dataset_skripsi.xlsx')
df.head()
df.shape
df.info()
df.columns
df.isnull().sum()
df.describe()
```

```python
df.sample(5, random_state=42)

df['text_length'] = df['ulasan'].astype(str).apply(len)
df['text_length'].describe()

plt.figure()
plt.hist(df['text_length'], bins=30)
plt.title('Distribusi Panjang Ulasan')
plt.xlabel('Panjang Teks')
plt.ylabel('Jumlah Ulasan')
plt.show()
```
