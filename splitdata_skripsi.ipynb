```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("/content/hasil_preprocessing.csv")
df.columns
```

```python
train_df, test_df = train_test_split(
    df,
    test_size=0.3,
    random_state=42,
    shuffle=True
)

print("Train size:", len(train_df))
print("Test size:", len(test_df))
```

```python
manual_sample = train_df.sample(
    n=350,
    random_state=42
)

manual_sample = train_df.sample(
    n=350,
    random_state=42
)
```

```python
manual_sample = manual_sample.copy()
manual_sample["aspek"] = ""
manual_sample["sentiment"] = ""

manual_sample.to_excel(
    "train_manual_label.xlsx",
    index=False
)

test_df.to_excel(
    "data_test_raw.xlsx",
    index=False
)
```

```python
print("Total data:", len(df))
print("Train:", len(train_df))
print("Test:", len(test_df))
print("Manual label:", len(manual_sample))
```
